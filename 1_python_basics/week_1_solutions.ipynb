```python
l = [2, 13, 18, 93, 22]

def func(lst):
    groups = ([], [])
    for num in lst:
        if num%2 == 0:
            groups[0].append(num)
        else:
            groups[1].append(num)
    return groups


even_list, odd_list = func(l)
```

```python
set1 = set(['data', 'python'])
set2 = set(['data', 'function', 'qcut', 'lambda', 'python', 'miuul'])


if set2.issuperset(set1):
    print(set2.intersection(set1))
```

```python
def control(set_1, set_2):
    if set_1.issuperset(set_2):
        print(set_1.intersection(set_2))
    else:
        print(set_2.difference(set_1))
        
```

```python
control(set1, set2)
```
