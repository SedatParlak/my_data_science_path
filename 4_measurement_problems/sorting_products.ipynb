```python
def weighted_sorting_score(dataframe, w1=32, w2=36, w3=42):
    
    total_score = df['purchase_count_scaler'] * w1 + \
    df['comment_count_scaler'] * w2 + \
    df['rating'] * w3
    
    return total_score / 100
```

# 4) Bayesian Average Rating Score

```python
def bayesian_average_rating_score(n, confidence=0.95):
    
    if sum(n) == 0:
        return 0
    
    K = len(n)
    z = st.norm.ppf(1 - (1 - confidence) / 2)
    N = sum(n)
    first_part = 0.0
    second_part = 0.0
    
    for k, n_k in enumerate(n):
        first_part += (k + 1) * (n[k] + 1) / (N + K)
        second_part += (k + 1) * (k + 1) * (n[k] + 1) / (N + K)
    score = first_part - z * math.sqrt((second_part - first_part * first_part) / (N + K +1))
    
    return score
```

**Calculate bayesian score**

```python
df['bayesian_score'] = df.apply(lambda x: bayesian_average_rating_score(x[['1_point',
                                                   '2_point',
                                                   '3_point',
                                                   '4_point',
                                                   '5_point']]), axis=1)
```

# 5) Hybrid Sorting

```python
def hybrid_sorting(dataframe, bay_weight=60, wss_weight=40):
    
    bay_score = dataframe.apply(lambda x: bayesian_average_rating_score(x[['1_point',
                                                   '2_point',
                                                   '3_point',
                                                   '4_point',
                                                   '5_point']]), axis=1)
    wss_score = weighted_sorting_score(dataframe)
    
    return (bay_score * bay_weight + wss_score * wss_weight) / 100
```

```python
df['hybrid_sorting_score'] = hybrid_sorting(df, bay_weight=60, wss_weight=40)
```

```python
df.sort_values('hybrid_sorting_score', ascending=False).head(10)
```
